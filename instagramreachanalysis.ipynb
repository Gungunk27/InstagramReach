# Instagram Reach Analysis using Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
data=pd.read_csv("Instagram data.csv",encoding="latin1")
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
data.head()
```

```python
data.info()
data.describe()
```

```python
plt.figure(figsize=(10,8))
plt.style.use("fivethirtyeight")
plt.title("Distribution of data from Home")
sns.histplot(data['From Home'])
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of data from Impressions")
sns.histplot(data["Impressions"])
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of data using Hashtags")
sns.histplot(data["From Hashtags"])
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of data using Explore")
sns.histplot(data["From Explore"])
plt.show()
```

```python
home=data['From Home'].sum()
hashtags=data['From Hashtags'].sum()
explore=data['From Explore'].sum()
others=data['From Other'].sum()

labels=['home','hashtags','explore','others']
values=[home,hashtags,explore,others]
fig=px.pie(data,values=values,labels=labels,title="Impression on Insta Post from various sources",hole=0.5)
fig.show()
```

```python
text=" ".join(i for i in data.Caption)
stopwords=set(STOPWORDS)
wordcloud=WordCloud(stopwords=stopwords,background_color="white").generate(text)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud,interpolation="bilinear")
```

```python
text=" ".join(i for i in data.Hashtags)
stopwords=set(STOPWORDS)
wordcloud=WordCloud(stopwords=stopwords,background_color="white").generate(text)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud,interpolation="bilinear")
```

```python
corr=data.corr()
print(corr["Impressions"].sort_values(ascending=False))
```

```python
rate=data["Follows"].sum()/data["Profile Visits"].sum()
print(rate)
```

```python
x=np.array(data[["Follows","Likes","Comments","Shares","Saves","Profile Visits"]])
y=np.array(data["Impressions"])
xtest,xtrain,ytest,ytrain=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=PassiveAggressiveRegressor()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```
